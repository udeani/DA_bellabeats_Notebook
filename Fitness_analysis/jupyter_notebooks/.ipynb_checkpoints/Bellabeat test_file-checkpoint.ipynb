#### checking if the user steps activity in minutes is equivalent to hourly or daily activity

### Minute-Daily steps check

```python
# for 7, 5 and 13 and 8

# daily steps
daily_steps = pd.read_csv(file_list[7])
print(daily_steps.shape)
daily_steps.head(3)
```

```python
# sorting the minute wide steps for grouping
minute_steps_wide = pd.read_csv(file_list[5])
minute_steps_wide["Date"] = minute_steps_wide["ActivityHour"].astype("string").str[:9]
#minute_steps_wide["Time"] = minute_steps_wide["ActivityHour"].astype("string").str[9:]
#minute_steps_wide["Date"] = pd.to_datetime(minute_steps_wide["Date"])

#summing up the steps by rows to get Column "Total"
minute_steps_wide["Total"] = minute_steps_wide.iloc[:, 2:-1].sum(axis=1)

#dropping the steps count columns
minute_steps_wide.drop(columns=minute_steps_wide.iloc[:,2:-2], inplace=True)

minute_steps_wide.head(3)
```

```python
# using pivot table to sum the steps by date and column wise
minute_steps_wide = pd.pivot_table(minute_steps_wide, index=["Id","Date"],values="Total",aggfunc=np.sum)

minute_steps_wide.head(3)
```

```python
#for the narrow minutes step
minute_steps_narrow = pd.read_csv(file_list[13])

#converting the datetime to date string for easy sorting
minute_steps_narrow["Date"] = minute_steps_narrow["ActivityMinute"].astype("string").str[:9]

#creating a pivot table for the Dataframe
minute_steps_narrow = pd.pivot_table(minute_steps_narrow, index=["Id","Date"], values="Steps", aggfunc=np.sum)

minute_steps_narrow.head(3)
```

```python
# Merging the narrrow and the wide data together(reseting the index to avoid error)
min_step_total = pd.merge(minute_steps_wide.reset_index(), minute_steps_narrow.reset_index(), on=["Id","Date"], how="right")

min_step_total.head(3)
```

```python
mins_daily_steps = pd.concat([min_step_total, daily_steps],axis=1)
mins_daily_steps
```

### Minute-Daily step Finding
From the above validations, it is clear that the daily_steps has more complete data than the minutes:

    1. missing Data in some dates
    2. incomplete rows

### Hourly-Daily steps check

```python
hourly_steps = pd.read_csv(file_list[8])

# making a date string for easy pivoting
hourly_steps["Date"] = hourly_steps["ActivityHour"].astype("string").str[:9]

#making a pivot
hourly_steps = pd.pivot_table(hourly_steps, index=["Id","Date"], values="StepTotal", aggfunc=np.sum)

#reseting the index for merging/concatetion with daily_steps
hourly_steps = hourly_steps.reset_index()

hourly_steps.head(3)
```

```python
# merging the hourly and the daily

hourly_daily_steps = pd.concat([hourly_steps, daily_steps], axis=1)

hourly_daily_steps
```

### Hourly-Daily steps finding
It is obvoius that the hourly and the minute_total are the same. but the fact that the daily is still more complete

### Validating the Calories (minute vs daily)

```python
# for the narrow data
min_calo_narrow = pd.read_csv(file_list[16])

# creating a date column
min_calo_narrow["Date"] = min_calo_narrow["ActivityMinute"].astype("string").str[:9]

# adding the steps by day using pivot table
min_calo_narrow = pd.pivot_table(min_calo_narrow, index=["Id","Date"], values="Calories", aggfunc=np.sum)

# reseting the index
min_calo_narrow.reset_index()

min_calo_narrow.head(3)
```

```python
min_calo_wide = pd.read_csv(file_list[3])

# creating a date string for easy sorting
min_calo_wide["Date"] = min_calo_wide["ActivityHour"].astype("string").str[:9]

# adding calories row wise
min_calo_wide["Total"] = min_calo_wide.iloc[:, 2:-1].sum(axis=1)

# Droping the less relavent columns
min_calo_wide.drop(columns=min_calo_wide.iloc[:, 2:-2], inplace=True)

# calculating the sum using pivot_table
min_calo_wide = pd.pivot_table(min_calo_wide, index=["Id", "Date"], values="Total", aggfunc=np.sum)
min_calo_wide.head(3)
```

```python
# merging the narrow and wide data of calories

min_calo_total = pd.merge(min_calo_narrow, min_calo_wide, on=["Id","Date"], how="right")
min_calo_total
```

```python
# daily calories
daily_calo = pd.read_csv(file_list[2])
print(daily_calo.shape)
daily_calo.head(3)
```

```python
# concatenating the minute and the daily for validation

min_daily_calo = pd.concat([min_calo_total.reset_index(), daily_calo.reset_index()], axis=1)

min_daily_calo
```

### Minute-Daily Calories Finding
